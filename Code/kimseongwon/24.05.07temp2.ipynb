```python
import plotly.graph_objects as go

# 데이터 생성
x = df_sort['temp_num']
y = df_sort['차이절대값합']

# 미분 진행
dx = x[1] - x[0]
dy = np.diff(y) / dx

# 그래프 그리기
fig = go.Figure(data=go.Scatter(x=x, y=y))

fig.add_trace(go.Scatter(x=x[:-1], y=dy, mode='lines', name='Derivative'))

fig.update_layout(width=800, height=600)

fig.show()
```

```python
copy = temp_df_work.copy()
```

```python
copy
```

```python
# 일단 이상치를 max 값으로 대체
copy['전체_건수'] = copy['전체_건수'].apply(lambda x: 58 if x > 58 else x)
```

```python
import plotly.express as px

fig = px.box(data_frame= copy, y="전체_건수")

fig.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(copy['전체_건수'])
plot_pacf(copy['전체_건수'])
plt.show()

model = ARIMA(copy['전체_건수'], order=(1, 1, 0))  # p, d, q는 선택한 파라미터입니다.
model_fit = model.fit()

# 예측 결과 출력
print(model_fit.summary())

# t = model_fit.predict(start=0, end=len(copy)-1, typ='levels')

plt.plot(copy['전체_건수'].index, copy['전체_건수'].values, label='Actual')
plt.plot(copy['전체_건수'].index, model_fit.predict(start=0, end=len(copy)-1), label='Forecast')
plt.title('Actual vs. Predicted Bike Counts')
plt.xlabel('Date')
plt.ylabel('Bike Count')
plt.legend()
plt.grid(True)
plt.show()

forecast = model_fit.forecast(steps=1)
# print("다음 시간대 예측값:", forecast[0])
print(type(forecast))
print(forecast)
```
