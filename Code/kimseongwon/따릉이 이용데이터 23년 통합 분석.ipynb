```python
raw = pd.read_csv(path, encoding='utf-8')

df = raw.copy()

df
```

```python
df_master = pd.read_csv('../../../../강남 따릉이_data/서울시 따릉이대여소 마스터 정보.csv', encoding='cp949')

df_master
```

```python
대여소_data = df_master[['대여소_ID', '주소1']]

대여소_data
```

```python
대여소_data['주소1'] = 대여소_data['주소1'].str.extract(r'(\S+)구')

대여소_data
```

```python
강남_대여소_ID = 대여소_data.loc[대여소_data['주소1'] == '강남', '대여소_ID']
```

```python
강남_대여소_ID
```

```python
condition1 = df['시작_대여소_ID'].isin(강남_대여소_ID)
condition2 = df['종료_대여소_ID'].isin(강남_대여소_ID)

강남_mask = df[condition1 & condition2]
강남_mask
```
