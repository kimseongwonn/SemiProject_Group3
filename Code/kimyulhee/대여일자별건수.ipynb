```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
file_path = 'C:/Users/evol0/OneDrive/바탕 화면/새 폴더 (2)/서울특별시-공공자전거-일별-대여건수_23.1-6.csv'

# CSV 파일 읽기
df = pd.read_csv(file_path, encoding='cp949')



# 선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(df['대여일시'], df['대여건수'], marker='o', linestyle='-')
plt.title('일별 대여건수')
plt.xlabel('날짜')
plt.ylabel('대여건수')
plt.grid(True)
plt.show()
```
