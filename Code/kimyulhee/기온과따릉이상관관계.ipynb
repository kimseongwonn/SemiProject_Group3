```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 파일 로드
temperature_data = pd.read_csv('../따릉이2/기온.csv', encoding='cp949')  # 일평균기온 데이터 파일
bicycle_data = pd.read_csv('../따릉이2/일별따릉이데이터.csv')  # 일 따릉이 자전거 대여량 데이터 파일

# 날짜 기준으로 데이터 병합
merged_data = pd.merge(temperature_data, bicycle_data, on='기준_날짜')

# 상관관계 계산
correlation = merged_data['평균기온(°C)'].corr(merged_data['전체_건수'])
print(f"일평균기온과 일 따릉이 자전거 대여량의 상관관계는 {correlation:.2f}입니다.")

# 산점도 그래프 생성
sns.scatterplot(x='평균기온(°C)', y='전체_건수', data=merged_data)
plt.title(f"평균기온(°C)과 전체_건수의 상관관계 (r={correlation:.2f})")
plt.xlabel('일평균기온 (°C)')
plt.ylabel('전체_건수')
plt.grid(True)
plt.show()
```
