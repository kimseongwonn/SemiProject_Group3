# 데이터 불러오기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm

font_path = 'C:\\Windows\\Fonts\\gulim.ttc'
font = fm.FontProperties(fname=font_path).get_name()
mpl.rc('font', family=font)

강서_대여소_ID = pd.read_csv('../../Data/따릉이/강서_따릉이_대여소_ID.csv',encoding='utf-8')

강서_데이터1 = pd.read_csv('../../Data/따릉이/강서_따릉이_데이터_01.csv',encoding='utf-8')
강서_데이터2 = pd.read_csv('../../Data/따릉이/강서_따릉이_데이터_02.csv',encoding='utf-8')

강서_데이터 = pd.concat([강서_데이터1, 강서_데이터2], ignore_index=True)
print(len(강서_데이터))
```

# 일별 대여데이터 시각화

```python
일별_데이터 = 강서_데이터.groupby('기준_날짜')['전체_건수'].sum()

눈금_간격 = 5  # 예시로 눈금 간격을 5로 설정
눈금_위치 = range(0, len(일별_데이터.index), 눈금_간격)
눈금_레이블 = 일별_데이터.index[::눈금_간격]

# 시작 대여소별 대여 건수 시각화
일별_데이터.plot(kind='bar')
plt.title('강서 지역 일별 대여 건수')
plt.xlabel('날짜')
plt.ylabel('대여 건수')
plt.xticks(rotation=90)
plt.xticks(눈금_위치, 눈금_레이블, rotation=90)
plt.show()
```

# 시간별 데이터 시각화
- 출근 시간 : 06:00~10:00
- 퇴근 시간 : 17:00~21:00

```python
def divide_time_range(time):
    if time < 600:
        return '00:00~06:00'
    elif time < 1000:
        return '06:00~10:00'
    elif time < 1600:
        return '10:00~17:00'
    elif time < 2000:
        return '17:00~21:00'
    else:
        return '20:00~24:00'

강서_데이터['시간대_구간'] = 강서_데이터['기준_시간대'].apply(divide_time_range)

# 시간대 구간별로 건수 합산
강서_시간대별_건수 = 강서_데이터.groupby('시간대_구간')['전체_건수'].sum()
```

```python
plt.figure(figsize=(12, 6))
강서_시간대별_건수.plot(kind='bar')
plt.title('시간대별 따릉이 확인')
plt.xlabel('대여소 ID')
plt.ylabel('대여 건수')
plt.xticks(rotation=90)
plt.show()
```

```python
# 시작 대여소 ID로 묶은 후 시간대별 건수 계산
강서_시작_대여소별_시간대별_대여량 = 강서_데이터.groupby(['시작_대여소_ID', '시간대_구간'])['전체_건수'].sum()

# 종료 대여소 ID로 묶은 후 시간대별 건수 계산
강서_종료_대여소별_시간대별_반납량 = 강서_데이터.groupby(['종료_대여소_ID', '시간대_구간'])['전체_건수'].sum()
```

```python
import matplotlib.pyplot as plt

# 출근 시간(06:00~10:00)에 해당하는 데이터 추출
출근시간_대여량 = 강서_시작_대여소별_시간대별_대여량.xs('06:00~10:00', level='시간대_구간')

# 퇴근 시간(17:00~21:00)에 해당하는 데이터 추출
퇴근시간_대여량 = 강서_시작_대여소별_시간대별_대여량.xs('17:00~21:00', level='시간대_구간')

# 출퇴근시간 대여량을 따로 그래프로 시각화
plt.figure(figsize=(12, 8))

# 출근시간은 막대 그래프로
plt.bar(출근시간_대여량.index, 출근시간_대여량.values, width=0.4, label='출근시간', color='skyblue')

# 퇴근시간은 선 그래프로
plt.plot(퇴근시간_대여량.index, 퇴근시간_대여량.values, label='퇴근시간', color='orange', marker='o')

plt.title('대여소별 출근 시간과 퇴근 시간 대여량 비교')
plt.xlabel('대여소 ID')
plt.ylabel('대여량')
plt.xticks(rotation=45)
plt.legend(title='시간대')
plt.tight_layout()
plt.show()
```

# 한시간 단위 시각화
 - 아직 사용처 안 나옴

```python
def divide_time_range1(time):
    for hour in range(24):
        if time < (hour + 1) * 100:
            return f'{hour:02d}:00~{(hour + 1):02d}:00'

강서_데이터['한시간_구간'] = 강서_데이터['기준_시간대'].apply(divide_time_range1)

# 시간대 구간별로 건수 합산
강서_시간대별_건수2 = 강서_데이터.groupby('한시간_구간')['전체_건수'].sum()
```

```python
plt.figure(figsize=(12, 6))
강서_시간대별_건수2.plot(kind='bar')
plt.title('시간대별 따릉이 확인')
plt.xlabel('대여소 ID')
plt.ylabel('대여 건수')
plt.xticks(rotation=90)
plt.show()
```

```python
강서_데이터['한시간_구간'] = 강서_데이터['기준_시간대'].apply(divide_time_range1)

# 시간대 구간별로 건수 합산
강서_시간대별_건수_시작 = 강서_데이터.groupby(['시작_대여소_ID','한시간_구간'])['전체_건수'].sum()
강서_시간대별_건수_종료 = 강서_데이터.groupby(['종료_대여소_ID','한시간_구간'])['전체_건수'].sum()
```

```python
import matplotlib.pyplot as plt

# 시작 대여소 ID로 묶은 후 시간대별 대여량 그래프
강서_시간대별_건수_시작.unstack().plot(kind='line', figsize=(10, 6), linewidth=2, marker='o')
plt.title('강서 지역 대여소별 시간대별 대여량')
plt.xlabel('대여소_ID')
plt.ylabel('대여량')
plt.legend(title='대여소 ID', bbox_to_anchor=(1, 1))
plt.tight_layout()  # 그래프 간격 조정
plt.show()

# 종료 대여소 ID로 묶은 후 시간대별 반납량 그래프
강서_시간대별_건수_종료.unstack().plot(kind='line', figsize=(10, 6), linewidth=2, marker='o')
plt.title('강서 지역 대여소별 시간대별 반납량')
plt.xlabel('대여소_ID')
plt.ylabel('반납량')
plt.legend(title='대여소 ID', bbox_to_anchor=(1, 1))
plt.tight_layout()  # 그래프 간격 조정
plt.show()
```

# 요일별 비교

```python
import matplotlib.pyplot as plt
import pandas as pd

# 강서_데이터가 날짜 열을 포함한 데이터프레임이라고 가정합니다.
# '기준_날짜' 열을 날짜 형식으로 변환합니다.
강서_데이터['기준_날짜'] = pd.to_datetime(강서_데이터['기준_날짜'], format='%Y-%m-%d')

# '기준_날짜' 열에서 요일을 추출하여 새 열을 추가합니다.
강서_데이터['요일'] = 강서_데이터['기준_날짜'].dt.day_name()

# 영어 요일명을 한글 요일명으로 변환하는 사전을 만듭니다.
영어_한글_요일_매핑 = {
    'Monday': '월',
    'Tuesday': '화',
    'Wednesday': '수',
    'Thursday': '목',
    'Friday': '금',
}

# 요일 순서를 정의합니다.
요일_순서 = ['월', '화', '수', '목', '금']

# '요일' 열의 값을 한글 요일명으로 변환합니다.
강서_데이터['요일'] = 강서_데이터['요일'].map(영어_한글_요일_매핑)

# 요일별 전체 건수 통계를 계산합니다.
요일별_통계 = 강서_데이터.groupby('요일')['전체_건수'].sum()

# 막대 그래프를 그립니다.
plt.figure(figsize=(8, 6))
요일별_통계[요일_순서].plot(kind='bar', color='skyblue')
plt.title('요일별 전체 건수')
plt.xlabel('요일')
plt.ylabel('대여 건수')
plt.xticks(rotation=90)  # x축 레이블 회전
plt.show()
```

# ARIMA
 - 시계열 데이터의 트랜드와 계절성을 설명하는데 좋음 
 - 데이터가 정상성을 보이지 않을 떄 사용하기 좋음   
 - ARIMA 모델은 AR(p),MA(q), 차분(d)의 세가지 요소로 구성 
 
자기 회기(AR) :  현재 값이 이전의 값에 의존하는 모델. 현재 시점의 데이터가 이전 시점의 데이터에 영향 받는다고 가정함
                AR 모델은  시계열 데이터의 현재 값 이전 값들의 선형조합으로 예측하는 모델

차분(I, Integration) : 차분은 시계열 데이터의 불규칙한 변동을 보정하는 방법.
                       차분 지수(d)는 데이터의 계절성 및 추세를 제거하는 데 사용
    
이동 평균(MA) : 이전 예측 오차를 보정하여 모델을 향상시키는데 사용.
               모델이 이전 예측에서 벗어난 정도를 평가하고 이를 다음 예측에 반영
               MA 모델은 예측 오차의 이동 평균을 사용하여 다음 값을 예측하는 모델

트랜드 : 시간에 따른 데이터의 장기적인 변화나 추세

계절성 : 연도의 특정 시기나 주기에 발생하는 패턴이나 주기성

정상성 : 정상성을 가진 시계열 데이터는 시간에 따라 평균과 분산이 일정하며, 시간에 관계없이 일정한 패턴을 보입


# SARIMA (Seasonal ARIMA)
- ARIMA 모델의 계절성 요소를 처리하기 위해 사용
- 시즌 주기를 고려하여 데이터를 모델링함

ARIMA에서 계절성 요소를 추가한거  : 자기 회기, 차분, 이동 평균은 동일
+ 계절성(S) : 특정한 계절 패턴이나 주기성을 모델링하는 요소. 계절성은 주어진 주기에 대해 반복되는 패턴을 의미

# 지수 평활법 
- 시계열 데이터 예측하는데 간단하면서도 효과적
- 시계열 데이터의 추세와 계절성을 고려하여 데이터의 평활화와 예측을 수행.
-  지수평활법은 단순 지수평활법(Simple Exponential Smoothing), Holt's Linear Trend Model, Holt-Winters Model. 대여 건수의 추세와 계절성을 예측

가중 평균(Weighted Average): 최근 데이터에 더 높은 가중치를 부여하여 이동평균을 계산합니다. 이는 더 최근의 데이터가 예측에 더 큰 영향을 미치도록 합니다.

지수평활계수(Exponential Smoothing Coefficient): 이 계수는 예측에 사용되는 최근 데이터의 가중치를 결정합니다. 일반적으로 0과 1 사이의 값으로 설정되며, 높은 값은 최근 데이터에 더 많은 가중치를 부여합니다.

# 지수평활법
- 시계열 데이터의 추세와 계절성을 고려하여 데이터의 평활화와 예측을 수행. 지수평활법은 단순 지수평활법(Simple Exponential Smoothing), Holt's Linear Trend Model, Holt-Winters Model. 대여 건수의 추세와 계절성을 예측

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Exponential Smoothing 모델을 초기화합니다.
model = ExponentialSmoothing(강서_데이터['전체_건수'], trend='add', seasonal='add', seasonal_periods=5)

# 모델을 학습합니다.
model_fit = model.fit()

# 모델로 대여 건수를 예측합니다.
예측값 = model_fit.predict(start=0, end=len(강서_데이터)-1)
```

```python
import matplotlib.pyplot as plt

# 예측 결과를 시각화합니다.
plt.plot(강서_데이터['기준_날짜'], 강서_데이터['전체_건수'], label='실제값')
plt.plot(강서_데이터['기준_날짜'], 예측값, label='예측값')
plt.legend()
plt.title('Exponential Smoothing 예측 결과')
plt.xlabel('날짜')
plt.ylabel('대여 건수')
plt.xticks(rotation=45)
plt.show()
```

# ARIMA
- RIMA 모델은 시계열 데이터의 자기상관성과 추세를 고려하여 예측하는 모델. ARIMA 모델은 대여 건수의 변동성과 추세를 파악하고 예측

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# '기준_날짜' 열을 날짜 형식으로 변환
강서_일별_데이터['기준_날짜'] = pd.to_datetime(강서_일별_데이터['기준_날짜'])

# ARIMA 모델을 초기화합니다.
model = ARIMA(강서_일별_데이터['전체_건수'], order=(5,1,0))

# 모델을 학습합니다.
model_fit = model.fit()


# 모델로 전체 일별 대여 건수를 예측합니다.
예측값 = model_fit.predict(start=0, end=len(강서_일별_데이터)-1, typ='levels')

# 전체 일별 대여 건수 시각화
plt.figure(figsize=(10, 6))
plt.bar(강서_일별_데이터['기준_날짜'], 강서_일별_데이터['전체_건수'], color='blue', label='실제값')
plt.plot(강서_일별_데이터['기준_날짜'], 예측값, marker='o', color='red', linestyle='dashed', label='예측값')
plt.title('전체 일별 대여 건수 패턴 (ARIMA)')
plt.xlabel('날짜')
plt.ylabel('대여 건수')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

# 대여소 별 반납, 대여 자료

```python
import matplotlib.pyplot as plt

# 시작 대여소별 전체 횟수 계산
시작대여소별_전체횟수 = 강서_데이터.groupby('시작_대여소_ID')['전체_건수'].sum()

# 종료 대여소별 전체 횟수 계산
종료대여소별_전체횟수 = 강서_데이터.groupby('종료_대여소_ID')['전체_건수'].sum()

```

```python

# 막대 그래프 그리기
plt.figure(figsize=(10, 6))

# 시작 대여소별 전체 횟수 막대 그래프
plt.subplot(2, 1, 1)
시작대여소별_전체횟수.plot(kind='bar', color='skyblue')
plt.title('시작 대여소별 전체 횟수')
plt.xlabel('대여소 ID')
plt.ylabel('전체 횟수')
plt.xticks(rotation=90)

# 종료 대여소별 전체 횟수 막대 그래프
plt.subplot(2, 1, 2)
종료대여소별_전체횟수.plot(kind='bar', color='lightgreen')
plt.title('종료 대여소별 전체 횟수')
plt.xlabel('대여소 ID')
plt.ylabel('전체 횟수')
plt.xticks(rotation=90)

plt.tight_layout()

plt.show()
```

```python
# 시작 대여소별 전체 횟수와 종료 대여소별 전체 횟수를 하나의 데이터프레임으로 병합
merged_data = pd.merge(시작대여소별_전체횟수, 종료대여소별_전체횟수, left_index=True, right_index=True, how='outer')

# 대여소 ID를 인덱스로 설정
merged_data.index.name = '대여소 ID'

# 시작 대여소별과 종료 대여소별 전체 횟수를 가지는 새로운 데이터프레임 생성
merged_data.columns = ['시작 대여소별 전체 횟수', '종료 대여소별 전체 횟수']


# 종료 대여소와 종료 대여소 간의 차이가 큰 상위 10개의 대여소 선택
대여_상위10대여소 = (merged_data['시작 대여소별 전체 횟수'] - merged_data['시작 대여소별 전체 횟수']).nlargest(10).index

# 대표 10개 대여소에 해당하는 데이터 추출
대표10대여소_데이터2 = merged_data.loc[대여_상위10대여소]

# 막대 그래프 그리기
plt.figure(figsize=(10, 6))

# 대표 10개 대여소에 대한 시작 및 종료 대여소별 전체 횟수 막대 그래프
대표10대여소_데이터2.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('대표적인 10개 대여소의 시작 및 종료 전체 횟수')
plt.xlabel('대여소 ID')
plt.ylabel('전체 횟수')
plt.xticks(rotation=90)
plt.legend(title='횟수 유형')

plt.tight_layout()
plt.show()
```
